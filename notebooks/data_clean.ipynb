```python
df = pd.read_csv('nba_player_totals_2023.csv')
```

```python
df
```

```python
#drop unnecessary columns
df = df.drop(df.columns[[0,1]], axis = 1)
```

```python
#add percentage columns
df['fg_percentage'] = round(df['made_field_goals'] / df['attempted_field_goals'], 3)
df['three_point_percentage'] = round(df['made_three_point_field_goals'] / df['attempted_three_point_field_goals'], 3)
df['free_throw_percentage'] = round(df['made_free_throws'] / df['attempted_free_throws'], 3)

#add 'per game' columns
df['apg'] = round(df['assists'] / df['games_played'], 1)
df['rpg'] = round((df['offensive_rebounds'] + df['defensive_rebounds']) / df['games_played'], 1)
df['spg'] = round(df['steals'] / df['games_played'], 1)
```

```python
df.head()
```

```python
columns = ['name',
 'positions',
 'age',
 'team',
 'games_played',
 'games_started',
 'minutes_played',
 'made_field_goals',
 'attempted_field_goals',
 'fg_percentage',
 'made_three_point_field_goals',
 'attempted_three_point_field_goals',
 'three_point_percentage',
 'made_free_throws',
 'attempted_free_throws',
 'free_throw_percentage',
 'offensive_rebounds',
 'defensive_rebounds',
 'rpg',
 'assists',
 'apg',
 'steals',
 'spg',
 'blocks',
 'turnovers',
 'personal_fouls',
 'points',
 'ppg']
```

```python
df = df[columns]
```

```python
df.head()
```

```python
df['ppg'] = round(df['ppg'], 1)
```
